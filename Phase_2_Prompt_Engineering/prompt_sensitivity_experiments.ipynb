```python
LECTURE_NOTES_1 = """
Course: Prompt Engineering for Generative AI
Lecture Topic: Prompt Sensitivity and Prompt Design

Large Language Models (LLMs) generate outputs based on patterns learned during pretraining. 
However, their responses are highly sensitive to the structure and phrasing of the prompt.

Prompt sensitivity refers to the phenomenon where small variations in instruction wording 
lead to noticeably different outputs. For example, asking a model to “summarize the text” 
versus “provide a concise summary of the key ideas” may result in different levels of detail 
or structure.

This sensitivity can cause unpredictability in real-world applications if prompts are not 
carefully engineered. As a result, prompt design must emphasize clarity, structure, and 
explicit constraints.

One approach to improving reliability is zero-shot prompting, where the model receives 
instructions without any examples. This relies entirely on the pretrained knowledge of the 
model. While zero-shot prompting is efficient, it often produces outputs with inconsistent 
formatting.

Few-shot prompting addresses this issue by providing demonstrations within the prompt. 
These examples show the model the expected reasoning pattern and output format. Even a 
single example can significantly improve consistency.

Another important technique is chain-of-thought prompting. This method encourages the 
model to generate intermediate reasoning steps before producing the final answer. Research 
has shown that chain-of-thought prompting improves performance on tasks that require 
multi-step reasoning.

Step-back prompting takes a slightly different approach by encouraging the model to first 
consider high-level principles before solving a specific problem. This allows the model to 
frame its reasoning within broader conceptual understanding.

Because LLM outputs may vary due to sampling randomness, self-consistency techniques are 
often used. In this approach, the system generates multiple candidate outputs and selects 
the most consistent or highest-quality answer.

In AI-powered educational tools, prompt engineering can be used to transform lecture 
content into interactive learning experiences. For example, lecture notes can be converted 
into diagnostic quizzes that help students actively retrieve information rather than 
passively rereading material.

Educational psychology research shows that retrieval practice improves long-term retention 
significantly compared to passive review strategies.

However, it is critical that generated educational content remains grounded in the source 
material. Prompts must therefore include strict instructions that prevent the model from 
introducing external facts or hallucinated information.

Overall, prompt engineering is a critical component in building reliable AI applications 
that produce structured, reproducible, and context-aware outputs.
"""
```

```python
def generate_quiz(notes, temperature=0.3):

    user_prompt = f"""
Create a diagnostic quiz based ONLY on the lecture notes below.

Requirements:
- 3 Multiple Choice Questions (Easy, Medium, Hard)
- 2 Short Answer Questions
- 1 Application Question
- Include answers and explanations

Lecture Notes:
{notes}
"""

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        temperature=temperature,
        messages=[
            {"role": "system", "content": SYSTEM_PROMPT},
            {"role": "user", "content": user_prompt}
        ]
    )

    return response.choices[0].message.content
```

```python
# Create different versions of the system prompt

SYSTEM_PROMPT_VARIANTS = {
    
    "baseline": SYSTEM_PROMPT,

    "variant_1": """
You are the AI Smart Learning Companion acting as a Socratic tutor for graduate students.

Rules:
- Use only the provided lecture notes
- Do not introduce external knowledge
- Maintain clear structure and consistency

Output:
3 MCQ (Easy, Medium, Hard)
2 Short Answer Questions
1 Application Question
Include answers and explanations.
""",

    "variant_2": """
You are a graduate teaching assistant generating a diagnostic quiz.

Constraints:
- Questions must be grounded strictly in the lecture notes
- Do not add external facts
- Maintain consistent academic formatting

Generate:
- Three MCQ labeled Easy, Medium, Hard
- Two Short Answer Questions
- One Application Question
Include answers and explanations.
"""
}

print("Prompt variants loaded:", len(SYSTEM_PROMPT_VARIANTS))
```

```python
# Temperature settings to test output variability

TEMPERATURES = [0.2, 0.5, 0.8]

print("Temperatures:", TEMPERATURES)
```

```python
def generate_quiz_with_prompt(system_prompt, notes, temperature=0.3):

    user_prompt = f"""
Create a diagnostic quiz based ONLY on the lecture notes below.

Requirements:
- 3 MCQ (Easy, Medium, Hard)
- 2 Short Answer Questions
- 1 Application Question
- Include answers and explanations

Lecture Notes:
{notes}
"""

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        temperature=temperature,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ]
    )

    return response.choices[0].message.content
```

```python
experiment_results = []

for prompt_name, prompt_text in SYSTEM_PROMPT_VARIANTS.items():
    
    for temp in TEMPERATURES:
        
        for i, notes in enumerate(DATASET, start=1):

            print("\n==============================")
            print("PROMPT:", prompt_name)
            print("TEMPERATURE:", temp)
            print("DATASET ITEM:", i)
            print("==============================\n")

            output = generate_quiz_with_prompt(prompt_text, notes, temperature=temp)

            print(output)
            print("\n" + "="*80 + "\n")

            experiment_results.append({
                "prompt": prompt_name,
                "temperature": temp,
                "dataset_item": i,
                "output": output
            })
```

```python
lecture_texts = []

for file in files:
    
    print("Reading:", file)
    
    reader = PdfReader(file)
    
    text = ""
    
    for page in reader.pages:
        if page.extract_text():
            text += page.extract_text()
    
    lecture_texts.append({
        "lecture": file,
        "text": text
    })

print("Lectures loaded:", len(lecture_texts))
```

```python
def chunk_text(text, chunk_size=1200):
    chunks = []
    
    for i in range(0, len(text), chunk_size):
        chunk = text[i:i + chunk_size]
        chunks.append(chunk)
    
    return chunks
```

```python
DATASET = []

for lecture in lecture_texts:
    lecture_name = lecture["lecture"]
    text = lecture["text"]
    
    chunks = chunk_text(text, chunk_size=1200)
    
    for i, chunk in enumerate(chunks, start=1):
        DATASET.append({
            "lecture": lecture_name,
            "chunk_id": f"{lecture_name}_chunk_{i}",
            "notes": chunk,
            "case_type": "typical"
        })

print("Typical dataset entries:", len(DATASET))
```

```python
def generate_quiz_baseline(notes, temperature=0.3):
    user_prompt = f"""
Create a diagnostic quiz based ONLY on the lecture notes below.

Requirements:
- 3 Multiple Choice Questions (Easy, Medium, Hard)
- 2 Short Answer Questions
- 1 Application Question
- Include answers and explanations
- Do NOT use outside knowledge

Lecture Notes:
{notes}
"""

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        temperature=temperature,
        messages=[
            {"role": "system", "content": HARDENED_SYSTEM_PROMPT},
            {"role": "user", "content": user_prompt}
        ]
    )

    return response.choices[0].message.content
```

```python
typical_docs = [item for item in DOCUMENTED_DATASET if item["case_type"] == "typical"]

print("Typical docs for RAG:", len(typical_docs))
```

```python
def get_embedding(text):
    response = client.embeddings.create(
        model="text-embedding-3-small",
        input=text
    )
    return response.data[0].embedding
    
```

```python
qdrant = QdrantClient(":memory:")

print("Qdrant ready")
```

```python
collection_name = "lecture_chunks"

qdrant.recreate_collection(
    collection_name=collection_name,
    vectors_config=VectorParams(size=1536, distance=Distance.COSINE)
)

print("Collection created")
```

```python
sample_docs = typical_docs[:40]

points = []

for item in sample_docs:
    embedding = get_embedding(item["notes"])

    points.append(
        PointStruct(
            id=str(uuid.uuid4()),
            vector=embedding,
            payload={
                "lecture": item["lecture"],
                "chunk_id": item["chunk_id"],
                "notes": item["notes"]
            }
        )
    )

qdrant.upsert(
    collection_name=collection_name,
    points=points
)

print("Uploaded points:", len(points))
```

```python
def retrieve_relevant_chunks(query, top_k=3):
    
    query_embedding = get_embedding(query)

    results = qdrant.query_points(
        collection_name=collection_name,
        query=query_embedding,
        limit=top_k
    )

    return results.points
```

```python
def generate_quiz_rag(query, top_k=3, temperature=0.3):

    results = retrieve_relevant_chunks(query, top_k=top_k)

    retrieved_notes = "\n\n".join([r.payload["notes"] for r in results])

    sources = "\n".join([
        f"{r.payload['lecture']} | {r.payload['chunk_id']}"
        for r in results
    ])

    user_prompt = f"""
Create a diagnostic quiz based ONLY on the retrieved lecture notes below.

Requirements:
- 3 Multiple Choice Questions (Easy, Medium, Hard)
- 2 Short Answer Questions
- 1 Application Question
- Include answers and explanations
- Do NOT use outside knowledge

Retrieved Lecture Notes:
{retrieved_notes}
"""

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        temperature=temperature,
        messages=[
            {"role": "system", "content": HARDENED_SYSTEM_PROMPT},
            {"role": "user", "content": user_prompt}
        ]
    )

    quiz_output = response.choices[0].message.content

    final_output = f"""
================ QUIZ ================

{quiz_output}

============= SOURCES USED =============

{sources}
"""

    return final_output
```

```python
def critique_system_prompt(system_prompt):
    critique_prompt = f"""
You are an expert prompt engineer reviewing a system prompt for an educational AI application.

Your task:
1. Identify strengths of the prompt
2. Identify weaknesses or ambiguities
3. Suggest concrete improvements
4. Focus on:
   - grounding in source notes
   - formatting consistency
   - resistance to hallucination
   - clarity of instructions
   - suitability for quiz generation

System Prompt to Review:
{system_prompt}

Return your answer in this format:
Strengths:
- ...

Weaknesses:
- ...

Suggested Improvements:
- ...
"""
    
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        temperature=0.2,
        messages=[
            {"role": "system", "content": "You are a precise and critical prompt engineering evaluator."},
            {"role": "user", "content": critique_prompt}
        ]
    )

    return response.choices[0].message.content
```

```python
prompt_critique = critique_system_prompt(BASELINE_PROMPT_STEP4)

print(prompt_critique)
```

```python
def improve_system_prompt(system_prompt, critique):
    improve_prompt = f"""
You are an expert prompt engineer.

Revise the following system prompt based on the critique provided.
Your goal is to produce an improved final system prompt for an AI quiz-generation assistant.

Requirements for the improved prompt:
- Must strictly ground outputs in lecture notes
- Must explicitly forbid outside knowledge
- Must enforce consistent quiz structure
- Must be clear, concise, and robust
- Must work well for graduate-level educational quiz generation

Original System Prompt:
{system_prompt}

Critique:
{critique}

Return ONLY the improved system prompt text.
"""
    
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        temperature=0.2,
        messages=[
            {"role": "system", "content": "You are an expert at rewriting system prompts for robustness and clarity."},
            {"role": "user", "content": improve_prompt}
        ]
    )

    return response.choices[0].message.content
```

```python
FINAL_SYSTEM_PROMPT = improve_system_prompt(BASELINE_PROMPT_STEP4, prompt_critique)

print(FINAL_SYSTEM_PROMPT)
```

```python
def generate_quiz_with_any_prompt(system_prompt, notes, temperature=0.3):
    user_prompt = f"""
Create a diagnostic quiz based ONLY on the lecture notes below.

Requirements:
- 3 Multiple Choice Questions (Easy, Medium, Hard)
- 2 Short Answer Questions
- 1 Application Question
- Include answers and explanations
- Do NOT use outside knowledge

Lecture Notes:
{notes}
"""

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        temperature=temperature,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ]
    )

    return response.choices[0].message.content
```

```python
import math
import statistics
```

```python
def generate_with_logprobs(system_prompt, notes, temperature=0.0):

    user_prompt = f"""
Create a diagnostic quiz based ONLY on the lecture notes below.

Requirements:
- 3 Multiple Choice Questions (Easy, Medium, Hard)
- 2 Short Answer Questions
- 1 Application Question
- Include answers and explanations
- Do NOT use outside knowledge

Lecture Notes:
{notes}
"""

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        temperature=temperature,
        logprobs=True,
        top_logprobs=0,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ]
    )

    text = response.choices[0].message.content
    token_items = response.choices[0].logprobs.content

    logprob_values = [item.logprob for item in token_items if item.logprob is not None]

    avg_neg_logprob = -sum(logprob_values) / len(logprob_values)
    perplexity = math.exp(avg_neg_logprob)

    return {
        "text": text,
        "perplexity": perplexity,
        "num_tokens": len(logprob_values)
    }
```

```python
perplexity_examples = [item for item in dev if item["case_type"] == "typical"][:5]

print("Perplexity eval examples:", len(perplexity_examples))
```

```python
perplexity_results = []

for item in perplexity_examples:

    baseline_result = generate_with_logprobs(BASELINE_PROMPT_STEP4, item["notes"], temperature=0.0)
    final_result = generate_with_logprobs(FINAL_SYSTEM_PROMPT, item["notes"], temperature=0.0)

    perplexity_results.append({
        "chunk_id": item["chunk_id"],
        "baseline_perplexity": baseline_result["perplexity"],
        "final_perplexity": final_result["perplexity"]
    })

print("Perplexity runs complete.")
```
